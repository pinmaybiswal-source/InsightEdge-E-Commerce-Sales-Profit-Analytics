```python
# Data manipulation and Data Loading Process
import pandas as pd
# encoding='latin-1': Special characters
data = pd.read_csv("C:/Users/pinma/Downloads/Sample - Superstore.csv",encoding='latin-1')
data
```

```python
# descriptive statistics of the dataset
data.describe()
```

```python
# Datasets information about null and datatype
data.info()
```

```python
# converting date Column
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

data.info()
```

```python
# Add column 
data['Order month'] = data['Order Date'].dt.month
data['Order year'] = data['Order Date'].dt.year
data['Order Day of week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

```python
# plotly.express (px): Data visualization library
import plotly.express as px 

#plotly.graph_objects (go): Advanced and customizable graphs
import plotly.graph_objects as go

#plotly.io (pio): Graph templates
import plotly.io as pio

# For Colors
import plotly.colors as colors

#pio.templates.default = "plotly_white": Default theme white
pio.templates.default = "plotly_white"
```

```python
# Monthly Analysis
sale_by_month = data.groupby('Order month')['Sales'].sum().reset_index()
fig = px.line(sale_by_month,
              x='Order month',
              y='Sales',
              title='Monthly Sales Analysis ')
fig.show()
```

```python
# Sales Analysis by Category
category_by_sale = data.groupby('Category')['Sales'].sum().reset_index()

# Create a pie chart
fig = px.pie(category_by_sale,
             names='Category',   # categories go here
             values='Sales',     # values to plot
             title='Category by Sales')
fig.show()
```

```python
# Sales analysis by sub-category
Sale_by_subcatagory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(Sale_by_subcatagory,
             x='Sub-Category',
             y='Sales',
             title='Sales By Sub-category')
fig.show()
```

```python
# Profit by Sales Analysis
profit_by_month = data.groupby('Order month')['Profit'].sum().reset_index()
fig = px.line(profit_by_month,x='Order month',
              y='Profit',
              title='Monthly Profit Analysis')
fig.show()
```

```python
# Analysis profit by catagory 
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
fig = px.pie(profit_by_category,
             names='Category',
             values='Profit',
             title='Profit Analysis By Catagory')
fig.show()
```

```python
# Analysis profit by sub-catagory 
profit_by_subcatagory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_subcatagory,
             x='Sub-Category',
             y='Profit',
             title='Profit Analysis By Subcategory')
fig.show()
```

```python
# Sales and Profit Analysis by Customer Segment
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], 
                     y=sales_profit_by_segment['Sales'], 
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], 
                     y=sales_profit_by_segment['Profit'], 
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()
```

```python
# Analyse sales-to-profit ratio
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
